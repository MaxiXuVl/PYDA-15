Исходные данные имеют следующую структуру:

```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
```

```
directories = {
```

### Пункт 1. Пользователь по команде "*p*" может узнать владельца документа по его номеру

Примеры работы:

1.

```
Введите команду:
p

Введите номер документа:
10006
```
Результат:  
`Владелец документа: Аристарх Павлов`

2.
```
Введите команду:
p

Введите номер документа:
12345
```
Результат:  
`Документ не найден в базе`

````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
````

````python
def com_p():
    number = input('Введите номер документа: ')
    res = 'Документ не найден в базе'
    for i in documents:
        if i['number'] == number:
            res = 'Владелец документа: ' + i['name']
    return res        

#com_p()
````

### Пункт 2. Пользователь по команде "*s*" может по номеру документа узнать на какой полке он хранится

Примеры работы:

1.

```
Введите команду:
s

Введите номер документа:
10006
```
Результат:  
`Документ хранится на полке: 2`

2.
```
Введите команду:
p

Введите номер документа:
12345
```
Результат:  
`Документ не найден в базе`

````python
def com_s(number):
    res = 'Документ не найден в базе'
    for i in directories.keys():
        if number in directories[i]:
            res = 'Документ хранится на полке: ' + i
    return res

#com_s('11-2')
````

### Пункт 3. Пользователь по команде "*l*" может увидеть полную информацию по всем документам

Пример работы:

```
Введите команду:
l
```

Результат:  
```
№: 2207 876234, тип: passport, владелец: Василий Гупкин, полка хранения: 1
№: 11-2, тип: invoice, владелец: Геннадий Покемонов, полка хранения: 1
№: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2
```

````python
def com_l():
    res = []
    for i in range(len(documents)):
        doc = documents[i]
        res.append('№: {}, тип: {}, владелец: {}, полка хранения: {}'.format(doc['number'], 
                                                                        doc['type'], 
                                                                        doc['name'], 
                                                                        com_s(number = doc['number'])))
    return res
#print(*com_l(), sep = '\n')
````

### Пункт 4. Пользователь по команде "*ads*" может добавить новую полку

Примеры работы:

1.

```
Введите команду:
ads

Введите номер полки:
10
```
Результат:  
`Полка добавлена. Текущий перечень полок: 1, 2, 3, 10.`

2.
```
Введите команду:
ads

Введите номер полки:
1
```
Результат:  
`Такая полка уже существует. Текущий перечень полок: 1, 2, 3.`

````python
#функция, которая возвращает список полок, каждый номер полки в котором имеет тип int
def racks(): 
    return str([int(i) for i in [*directories]]).strip('[]')
````

````python
def com_ads():
    number = input('Введите номер полки: ')
    if str(number) in directories.keys():
        res = f'Такая полка уже существует. Текущий перечень полок: {racks()}.'
        return res
    else:
        directories.setdefault(str(number), [])
        res = f'Полка добавлена. Текущий перечень полок: {racks()}.'
        return res

#com_ads()
````

### Пункт 5. Пользователь по команде "*ds*" может удалить существующую полку из данных (только если она пустая)

Примеры работы:

1.

```
Введите команду:
ds

Введите номер полки:
3
```
Результат:  
`Полка удалена. Текущий перечень полок: 1, 2.`

2.
```
Введите команду:
ds

Введите номер полки:
1
```
Результат:  
`На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: 1, 2, 3.`

3.
```
Введите команду:
ds

Введите номер полки:
4
```
Результат:  
`Такой полки не существует. Текущий перечень полок: 1, 2, 3.`

````python
def com_ds():
    number = input('Введите номер полки: ')
    if number in directories.keys():
        if len(directories[number]) == 0:
            directories.pop(number)
            res = f'Полка удалена. Текущий перечень полок: {racks()}.'
        else: 
            res = f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {racks()}.'
    else:
        res = f'Такой полки не существует. Текущий перечень полок: {racks()}.'
    return res
    
#com_ds()
````

````python
def distributor(com):
    if com == 'p':     #возвращает имя владельца по номеру документа
        print(com_p())
    elif com == 's':   #возвращает номер полки по номеру документа
        number = input('Введите номер документа: ')  ##здесь требует отдельно задать ввод, так как внутри функции эта команда не предусмотрена
        print(com_s(number))
    elif com == 'l':   #выводит всю информацию по документам
        print(*com_l(), sep = '\n')              ##здесь нужен спец вывод по элементам списка
    elif com == 'ads': #добавляет пустую полку, если ее нет 
        print(com_ads())
    elif com == 'ds':  #удаляет полку, если она пустая
        print(com_ds())
    else:
        print('Команда не найдена в базе. Попробуйте еще раз.')   
````

````python
#Итоговый код, в котором совмещаются все функции
com = input('Введите команду: ')
while com != 'q':
    distributor(com)
    com = input('Введите команду: ')
````

### Пункт 2. Пользователь по команде "*d*" может удалить документ из данных

```
d
```

````python
def com_d():
    number = input('Введите номер документа: ')
    list_numbers = {documents[i]['number']: i for i in range(len(documents))}
    if number in list_numbers:
        documents.pop(list_numbers[number])
        rack = com_s(number).split()[-1]
        directories[rack].remove(number)
        print('Документ удален.\nТекущий список документов:', *com_l(), sep = '\n')
    else:
        print('Документ не найден в базе.\nТекущий список документов:', *com_l(), sep = '\n')

#com_d()
````

````python
def com_m():
    number = input('Введите номер документа: ')
    rack_new = input('Введите номер полки: ')
    if number in [el['number'] for el in documents]:
        if rack_new in directories.keys():
            rack_old = com_s(number).split()[-1]
            directories[rack_old].remove(number)
            directories[rack_new].append(number)
            print('Документ перемещен.\nТекущий список документов:', *com_l(), sep = '\n')
        else:
            print(f'Такой полки не существует. Текущий перечень полок: {racks()}.')
    else:
        print('Документ не найден в базе.\nТекущий список документов:', *com_l(), sep = '\n')

#com_m()
````

````python
def distributor_all(com):
    if com == 'p':      #возвращает имя владельца по номеру документа
        print(com_p())
    elif com == 's':    #возвращает номер полки по номеру документа
        number = input('Введите номер документа: ')     #здесь требует отдельно задать ввод, так как внутри функции эта команда не предусмотрена
        print(com_s(number))
    elif com == 'l':    #выводит всю информацию по документам
        print(*com_l(), sep = '\n')                 #здесь нужен спец вывод по элементам списка
    elif com == 'ads':  #добавляет пустую полку, если ее нет 
        print(com_ads())
    elif com == 'ds':   #удаляет полку, если она пустая
        print(com_ds())
    elif com == 'ad':   #добавляет документ в данные, если указана существующая полка
        com_ad()        ##ничего не возвращает, просто печатает текущий перечень файлов с выполненным действием (добавление дока или нет)
    elif com == 'd':    #удаляет документ из данных
        com_d()         #ничего не возвращает, просто печатает текущий перечень файлов с выполненным действием (удалением или нет)
    elif com == 'm':    #перемещает док с одной полки на другую
        com_m()         #ничего не возвращает, просто печатает текущий перечень файлов с выполненным действием (перемещением дока или нет)
    else:
        print('Команда не найдена в базе. Попробуйте еще раз.')   
````
